```python
# Fonction pour lire les entités nommées d'un fichier CSV
def read_entities(file_path, column, max_tokens=6500):
    # entities = set()
    entities = []
    total_tokens = 0

    with open(file_path, 'r', encoding='utf-8') as file:
        reader = csv.DictReader(file, delimiter=';')
        for row in reader:
            if total_tokens >= max_tokens:
                break
            entity = row[column]
            if entity:
                # entities.add(row['Token'])
                entities.append(row['Token'])
            total_tokens += 1

    return entities

def stocker(chemin, contenu):
    w = open(chemin, "w")
    w.write(json.dumps(contenu, indent=2))
    w.close()
    # print(chemin)
    return chemin

# Fonction pour créer un UpSet plot
def create_upset_plot(data, output_path):
    memberships = from_contents(data)
    upset = UpSet(memberships, subset_size='count', show_percentages=True)
    upset.plot()
    plt.savefig(output_path)
    plt.show()

# Fonction pour traiter les fichiers et créer les UpSet plots par auteur
def process_files_by_author(author, annotation_dir, gold_dir, accord_dir, output_dir):
    
    tools = ['TesseractFra-PNG', 'Kraken-base',"REF"]

    
    # Lire les entités nommées pour gold et accord
    gold_files = {tool: f"resultat_Adjudication_{author}_petit-chose_{tool}_spaCy3.7.2-lg.csv.aia.csv" if author == 'DAUDET' else f"resultat_Adjudication_{author}_petite-Jeanne_{tool}_spaCy3.7.2-lg.csv.aia.csv" for tool in tools}
    accord_files = {tool: f"resultat_Adjudication_{author}_petit-chose_{tool}_spaCy3.7.2-lg.csv.aia.csv" if author == 'DAUDET' else f"resultat_Adjudication_{author}_petite-Jeanne_{tool}_spaCy3.7.2-lg.csv.aia.csv" for tool in tools}

    gold_entities = {tool: read_entities(os.path.join(gold_dir, author, file), 'gold', max_tokens) for tool, file in gold_files.items()}
    accord_entities = {tool: read_entities(os.path.join(accord_dir, author, file), 'accord', max_tokens) for tool, file in accord_files.items()}
    # return gold_entities

    # Lire les entités nommées pour les annotations
    annotation_files = {tool: f"CF_{author}_petit-chose_{tool}_spaCy3.7.2-lg.csv" if author == 'DAUDET' else f"CF_{author}_petite-Jeanne_{tool}_spaCy3.7.2-lg.csv" for tool in tools}
    annotation_entities = {tool: read_entities(os.path.join(annotation_dir, author, file), 'Label', max_tokens) for tool, file in annotation_files.items()}

    # Créer les dictionnaires pour UpSet plots par auteur
    data_gold = {'Gold -- Réf.': gold_entities['REF'],'Gold -- Tess. fr': gold_entities['TesseractFra-PNG'],'Gold -- Kraken': gold_entities['Kraken-base'], 'spaCy3.7 -- Réf.': annotation_entities['REF'], 'spaCy3.7 -- Tess. fr': annotation_entities['TesseractFra-PNG'], 'spaCy3.7 -- Kraken': annotation_entities['Kraken-base']}
    data_accord = {'Acc. MAJtr -- Réf.': accord_entities['REF'], 'Acc. MAJtr -- Kraken': accord_entities['Kraken-base'],'Acc. MAJtr -- Tess': accord_entities['TesseractFra-PNG'], 'spaCy3.7 -- Réf.': annotation_entities['REF'], 'spaCy3.7 -- Tess.fr': annotation_entities['TesseractFra-PNG'], 'spaCy3.7 -- Kraken': annotation_entities['Kraken-base']}

    # Créer les UpSet plots par auteur
    # output_path_gold = os.path.join(output_dir, f"{author}_gold_upset_plot.png")
    # create_upset_plot(data_gold, output_path_gold)
    # 
    # output_path_accord = os.path.join(output_dir, f"{author}_accord_upset_plot.png")
    # create_upset_plot(data_accord, output_path_accord)
    return data_gold
# Fonction pour traiter les fichiers et créer les UpSet plots globaux


# Appel à la fonction pour traiter les fichiers et générer les UpSet plots par auteur
authors = ['DAUDET', 'CARRAUD']
for auteur in authors:
    dico=process_files_by_author(auteur, annotation_dir, gold_dir, accord_dir, output_dir)
    print(dico)
# print(annotation_dir)
# print(dico.keys())
# print(dico["AccordMAJtr_Kraken"])
#     stocker(f"{auteur}_Dico_AccordMAJtr.json",dico)
    stocker(f"{auteur}_Dico_GOLD.json",dico)
```

```python
def process_files_globally(annotation_dir, gold_dir, accord_dir, output_dir):
    authors = ['DAUDET', 'CARRAUD']
    tools = ['TesseractFra-PNG', 'Kraken-base']

    global_gold_entities = {'Tess': set(), 'Kraken': set(), 'Gold': set()}
    global_accord_entities = {'Tess': set(), 'Kraken': set(), 'AccordMAJtr': set()}

    for author in authors:
        # Lire les entités nommées pour gold et accord
        gold_files = {tool: f"resultat_Adjudication_{author}_petit-chose_{tool}_spaCy3.7.2-lg.csv.aia.csv" if author == 'DAUDET' else f"resultat_Adjudication_{author}_petite-Jeanne_{tool}_spaCy3.7.2-lg.csv.aia.csv" for tool in tools}
        accord_files = {tool: f"resultat_Adjudication_{author}_petit-chose_{tool}_spaCy3.7.2-lg.csv.aia.csv" if author == 'DAUDET' else f"resultat_Adjudication_{author}_petite-Jeanne_{tool}_spaCy3.7.2-lg.csv.aia.csv" for tool in tools}

        gold_entities = {tool: read_entities(os.path.join(gold_dir, author, file), 'gold', max_tokens) for tool, file in gold_files.items()}
        accord_entities = {tool: read_entities(os.path.join(accord_dir, author, file), 'accord', max_tokens) for tool, file in accord_files.items()}

        # Lire les entités nommées pour les annotations
        annotation_files = {tool: f"CF_{author}_petit-chose_{tool}_spaCy3.7.2-lg.csv" if author == 'DAUDET' else f"CF_{author}_petite-Jeanne_{tool}_spaCy3.7.2-lg.csv" for tool in tools}
        annotation_entities = {tool: read_entities(os.path.join(annotation_dir, author, file), 'Label', max_tokens) for tool, file in annotation_files.items()}

        # Fusionner les entités pour chaque auteur
        for tool in tools:
            global_gold_entities['Tess'].update(annotation_entities['TesseractFra-PNG'])
            global_gold_entities['Kraken'].update(annotation_entities['Kraken-base'])
            global_gold_entities['Gold'].update(gold_entities[tool])

            global_accord_entities['Tess'].update(annotation_entities['TesseractFra-PNG'])
            global_accord_entities['Kraken'].update(annotation_entities['Kraken-base'])
            global_accord_entities['AccordMAJtr'].update(accord_entities[tool])

    # Créer les UpSet plots globaux
    global_output_path_gold = os.path.join(output_dir, "global_gold_upset_plot.png")
    create_upset_plot(global_gold_entities, global_output_path_gold)

    global_output_path_accord = os.path.join(output_dir, "global_accord_upset_plot.png")
    create_upset_plot(global_accord_entities, global_output_path_accord)

# Appel à la fonction pour traiter les fichiers et générer les UpSet plots globaux
# process_files_globally(annotation_dir, gold_dir, accord_dir, output_dir)
```
